```python
# Visualization Of Ouputs In The Form of Figure

comparePrice = data[['Min Price', 'Max Price', 'Modal Price']]
sns.boxplot(data=comparePrice)
plt.xlabel('Price Type')
plt.ylabel('Price')
plt.title('Price Distribution')
plt.show()
```

```python
sns.countplot(data=data, x='State')
plt.xlabel('State')
plt.ylabel('Count')
plt.title('Number of crops')
plt.xticks(rotation=90)
plt.show()
```

```python
data[data['State']=='Uttar Pradesh'].Commodity.unique()
```

```python
# Let's find Largest model prices collection of states
```

```python
state_num = set([x for x in data['State']])
state_num
```

```python
state_model_price={}
state_num = sorted(state_num)
for i in state_num:
    df = data[data['State']==i]
    
    state_model_price[i]= df['Modal Price'].sum()
    
state_model_price
```

```python
for i in state_num:
    print(i,"---->",state_model_price[i])
```

```python
fig = plt.figure(figsize =(10, 7))
plt.pie(state_model_price.values(), labels = state_num,rotatelabels=90)
```

```python
top_10_highest = pd.DataFrame(data.nlargest(10, 'Modal Price'))
top_10_lowest = pd.DataFrame(data.nsmallest(10, 'Modal Price'))
```

```python
top_10_highest
```

```python
plt.bar(top_10_highest["State"],top_10_highest['Max Price'])
plt.xlabel('State')
plt.ylabel('Max Price')
```

```python
# After a detail analysis of top 10 highest model prices ,we conclude that
# Kerala is best market for Cardamoms and Cloves .Also minimum and maximum price is also good. Seller can easily earn profit if they sell our Cardamons and Cloves kerala.

# Gujarat is best market for Cummin Seed(Jeera) .Also minimum and maximum price is also good. Seller can easily earn profit if they sell our Cummin Seed(Jeera) in Gujarat.

# NCT of Delhi is best market for Fish .Also minimum and maximum price is also good. It is profitable business in NCT of Delhi .
```

```python
top_10_lowest
```

```python
plt.bar(top_10_lowest["State"],top_10_lowest['Min Price'])
plt.xlabel('State')
plt.ylabel('Minimum Price')
```

```python
# After a detail analysis of top 10 lowest model prices ,we conclude that
# NCT of Delhi is not a good market for Flowers .Also minimum and maximum price is also not good. It is unprofitable business in NCT of Delhi .
```
