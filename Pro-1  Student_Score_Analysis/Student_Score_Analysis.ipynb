```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Expanded_data_with_more_features.csv")

# Staring 5 data/Info

print(df.head())
```

```python
# Data Types per Col and Row All information Memory,values etc

# Non null values

df.info()
```

```python
# Total no.of Null Values

# Null Value/isnull

df.isnull().sum()
```

```python
# Drop Unamed Column

# In Above OUTPUT

df = df.drop("Unnamed: 0",axis = 1)  # drop Unnamed Column
print(df.head())
```

```python
# CHANGE WEEKLY STUDY HR'S COL'N

df["WklyStudyHours"] = df["WklyStudyHours"].str.replace("5 - 10","5 - 10")    # it is in above o/p

df.head()
```

```python
# GENDER DISTRIBUTION

plt.figure(figsize = (5,5))

ax = sns.countplot(data = df, x = "Gender")
ax.bar_label(ax.containers[0])

plt.title("GENDER DISTRIBUTION")

plt.show()
```

```python
# Above Data Female > Male
```

```python
# Group by

# Mean Value[read,write,etc]---> .agg----->agrigate
gb = df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":"mean","WritingScore":"mean"})  

print(gb)
```

```python
plt.figure(figsize = (3,3))

sns.heatmap(gb)
plt.show()
```

```python
plt.figure(figsize = (3,3))

sns.heatmap(gb,annot = True)

plt.title("RelationShip B/W Parent's Edu and Student Score")

plt.show()
```

```python
# From above chart EDUCATION of Parent impact Student
```

```python
# Marage Status Impact on Student Score

gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":"mean","WritingScore":"mean"})  

print(gb1)
```

```python
# Fig  

plt.figure(figsize = (3,3))

sns.heatmap(gb1,annot = True)

plt.title("RelationShip B/W Parent's ParentMaritalStatus and Student Score")

plt.show()
```

```python
# ParentMaritalStatus No Impact For Student Score
```
